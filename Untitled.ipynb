```python
import pandas as pd
import matplotlib.pyplot as plt



# Read the CSV files
data_2021 = pd.read_csv("Tokyo 2021 dataset v3.csv")
data_2024 = pd.read_csv("Tokyo 2024.csv")

# Display the first few rows of each dataset to understand their structure
print(data_2021.head())
print(data_2024.head())

# Top 10 countries by total medals in 2021
top_10_2021 = data_2021.nlargest(10, 'Total')

# Plot the medal distribution
fig, ax = plt.subplots(figsize=(12, 6))
top_10_2021.plot(kind='bar', x='Team/NOC', y=['Gold Medal', 'Silver Medal', 'Bronze Medal'], stacked=True, ax=ax)

ax.set_title('Top 10 Countries by Total Medals in Tokyo 2021 Olympics')
ax.set_xlabel('Country')
ax.set_ylabel('Number of Medals')
plt.xticks(rotation=45)
plt.legend(title='Medal Type')
plt.tight_layout()
plt.show()

# Summary of sports and the number of events in the 2024 dataset
sports_summary_2024 = data_2024['sport'].value_counts()

# Plot the sports summary
fig, ax = plt.subplots(figsize=(12, 8))
sports_summary_2024.plot(kind='bar', ax=ax)

ax.set_title('Number of Events by Sport in Paris 2024 Olympics')
ax.set_xlabel('Sport')
ax.set_ylabel('Number of Events')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Top sports by number of events in 2024
top_sports_2024 = sports_summary_2024.nlargest(10)

# Display top sports by number of events in 2024
print(top_sports_2024)

# Predicting likely winners for 2024 based on 2021 data

# Normalizing the medal counts for each type of medal
data_2021['Gold Weighted'] = data_2021['Gold Medal'] * 3
data_2021['Silver Weighted'] = data_2021['Silver Medal'] * 2
data_2021['Bronze Weighted'] = data_2021['Bronze Medal'] * 1
data_2021['Total Weighted'] = data_2021['Gold Weighted'] + data_2021['Silver Weighted'] + data_2021['Bronze Weighted']

# Calculate the percentage of total weighted medals each country has
data_2021['Percentage'] = data_2021['Total Weighted'] / data_2021['Total Weighted'].sum()

# Calculate expected medal count for 2024 based on percentage
# Assume the total number of medals in 2024 will be the same as in 2021
total_medals_2021 = data_2021['Total'].sum()
data_2021['Expected Total Medals 2024'] = data_2021['Percentage'] * total_medals_2021

# Sorting based on expected total medals for 2024
predicted_winners_2024 = data_2021[['Team/NOC', 'Expected Total Medals 2024']].sort_values(by='Expected Total Medals 2024', ascending=False)

# Display the top 10 predicted winners for 2024
print(predicted_winners_2024.head(10))

# Plotting the predicted top 10 winners for 2024
fig, ax = plt.subplots(figsize=(12, 6))
predicted_winners_2024.head(10).plot(kind='bar', x='Team/NOC', y='Expected Total Medals 2024', ax=ax, color='skyblue')

ax.set_title('Top 10 Predicted Countries by Total Medals in Paris 2024 Olympics')
ax.set_xlabel('Country')
ax.set_ylabel('Expected Number of Medals')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
